# Sentinel API

* Connect to Sentinel API
* Open rift valley geojson and find the centroid of the polygon described in the geojson
* Paste the Geojson onto a folium map
* Query the API for our selected area
* Sort the data by ingestion date
* Run throught the data, downloading every zip file
* Unzip and remove the original zip file
* Plot the TCI image

The Sentinel API is a bit unreliable and I think they only let you query it so many times.

```python
import pandas as pd
import rasterio as rio
import matplotlib.pyplot as plt

import folium #I've heard of folium before its a good mapping tool
import geojson
import os
import time
import shutil
import math

from datetime import date
from sentinelsat import SentinelAPI, read_geojson, geojson_to_wkt
from pathlib import Path
from zipfile import ZipFile
from rasterio import plot
from collections import OrderedDict
from IPython.display import Image

data_folder = Path("../satellite_data/")
geojson_data = Path("../geojson/")
supporting_images = Path("../img")
```

```python
username = "cmeikle" #PLACE_YOUR_USERNAME_HERE
password = "Madone11"#PLACE_YOUR_PASSWORD_HERE
api = SentinelAPI(username, password, 'https://scihub.copernicus.eu/dhus')
```

Geojson got from https://geojson.io/#map=2/20.0/0.0 - very good and very simple to use

```python
geojson_file = geojson_data.joinpath("rift_valley_kenya.geojson")
with open(geojson_file) as f:
    gj = geojson.load(f)
polygon_points = gj[0]["geometry"]["coordinates"]
```

```python
# Finding the centoid of a polygon, using the "Of a finite set of points" method
# Remove final point as it is same as first
lat = 0
long = 0
for point in polygon_points[0]:
    lat += point[1]
    long += point[0]
    
no_points = len(polygon_points[0])
centroid = lat/no_points, long/no_points
# TODO check that this works with a wild polygon
```

```python
# Load the area we are interested in into folium to get a good instructural map
m = folium.Map(list(centroid))
folium.GeoJson(str(geojson_file)).add_to(m)
m
```

```python
# Load in the geojson footprint we are looking at
footprint = geojson_to_wkt(read_geojson(geojson_file))
# and query the api (there are other APIs available I believe Google Earth Engine, Landsat, planet)
products = api.query(footprint,
                     date = (date(2019, 6, 1), date(2020, 6, 1)),
                     platformname = 'Sentinel-2',
                     processinglevel = 'Level-2A',
                     cloudcoverpercentage = (0,3)
                    )
```

```python
#Pandas stuff

# Load all possible satellite datasets into a pandas dataframe, 
#geopandas can be used but I am getting an error when downloading
products_df = api.to_dataframe(products)
products_df.dropna(subset=["level1cpdiidentifier"], inplace=True)
# Sort in date order
products_df_sorted = products_df.sort_values(['ingestiondate'], ascending=[True])

# Create a date column
products_df_sorted["Date"] = products_df_sorted["ingestiondate"].dt.date

# Get evenly spread datasets from this date range, I started off with 6 and may start to increase it
occurences = round(len(products_df_sorted)/7)
products_df_sorted.reset_index(inplace=True)
products_df_sorted = products_df_sorted.loc[::occurences,:]
products_df_sorted.set_index("index", inplace=True)
```

```python
products_df_sorted
```

```python
image_dict = OrderedDict()
date_dict = OrderedDict()
#Unzip the files you have just downloaded
for index, filepath in products_df_sorted["title"].items():
    # download datasets (Can only do a few at a time)
    api.download(index, data_folder)
    # Define the filepath
    file_name = data_folder.joinpath(filepath + '.zip')
    # opening the zip file in READ mode
    try:
        with ZipFile(file_name, 'r') as _zip: 
            # extracting all the files
            print(f"Extracting {file_name}...")
            _zip.extractall(path=data_folder)
            print("Done!")
    except FileNotFoundError:
        print(f"The zip file hasn't been downloaded, seems to be some problem where the data isn't availale...")
        # Sleep to make sure the API doesn't get overloaded (Can only make so many queries in a certain time frame)
        time.sleep(10)
        continue
    # Remove the zip file once it has been unzipped
    if os.path.exists(file_name):
        os.remove(file_name)
    else:
        print(f"The file:{file} does not exist")
    
    # Look into the image folder to find the R10m tci image
    directory = data_folder.joinpath(filepath + ".SAFE")
    for root, dirs, files in os.walk(directory, topdown=False):
        if ("L2A" in root) and ("R10m" in root):
            for file in files:
                if "TCI" in file:
                    tci_r10_image_file = f"{root}/{file}" #true/total colour image
                    image = data_folder.joinpath(file)
                    break
                    
    try:
        # Move the R10m tci image from the IMG_DATA directory to our data_folder
        shutil.move(tci_r10_image_file, data_folder)
    except:
        print("The image already exists in the data folder")
    
    # Open the image using rasterio and append the image file to the dictionary
    image_dict[index] = rio.open(image)
    date_dict[index] = products_df_sorted.loc[index, "Date"]
    
    # Delete the .SAFE file to clear up space in memory
    if os.path.exists(directory):
        shutil.rmtree(directory)
    else:
        print(f"The directory:{directory} does not exist")
    
    """
    There may be a lot more interesting info as part of the sentinel 2 download, so I will have to pull that down too
    Information like the exact coordinates of the image. This can possibly be done in geopandas using crs?
    But I'm not sompletely familiar with geopandas and its differences with pandas
    """
    
    

```

```python
def figure_dimensions(number):
    # Find a list of factors of a number
    factors = []
    for i in range(1, number+1):
        if number % i == 0:
            factors.append(i)
            
    #find the middle two
    down = math.floor((len(factors)+1)/2) - 1 # take 1 away for the python indexing
    if len(factors) % 2 != 0:
        across = down
    else:
        across = down + 1
    return factors[down], factors[across]
```

```python
indexes = [k for k in image_dict.keys()]
down, across = figure_dimensions(len(indexes))
# Create a figure plot using matplotlib
fig, subplots = plt.subplots(down, across, figsize=(21,7))
subplots = subplots.flatten()
for number, image in enumerate(subplots):
    plot.show(image_dict[indexes[number]], ax=image)
    image.set_title(date_dict[indexes[number]])
```
